```python
import pyaudio
import wave
import socket
import os
import time
```

```python
CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNEL = 1
RATE = 44100
TIME = 60
```

```python
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
address = ('localhost', 5000)
s.bind(address)
s.listen(5)
print('ready to transmit BUT no one connect to me yet.')

```

```python
p = pyaudio.PyAudio()
wf = wave.open("sample1.wav",'rb')
stream = p.open(format = FORMAT,
                channels= CHANNEL,
                rate = RATE,
                output = True,
                frames_per_buffer = CHUNK)
data=wf.readframes(CHUNK)
print('stream ready')
```

```python
conn,addr=s.accept()
print('get a connect fromï¼š', addr)
time.sleep(5)
i = 0;
while(data != ''):
    #stream.write(data)
    data = wf.readframes(CHUNK)
    conn.send(data)
    print("Already transmit: " + str(i) + " frame")
    
print("transmit finished")
```

```python
stream.close()
p.terminate()
```
